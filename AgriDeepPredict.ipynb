```python

import pandas as pd

df = pd.read_csv('/content/FAOSTAT_data_en_1-8-2025 (3).csv')
def convert_usd_to_slc(row):
    if row['Unit'] == 'USD':
        row['Unit'] = 'SLC'
        y = row['Year']
        val = 0

        if y == 1991:
            val = 22.74
        elif y == 1992:
            val = 25.92
        elif y == 1993:
            val = 30.49
        elif y == 1994:
            val = 31.37
        elif y == 1995:
            val = 32.43
        elif y == 1996:
            val = 35.43
        elif y == 1997:
            val = 36.31
        elif y == 1998:
            val = 41.26
        elif y == 1999:
            val = 43.06
        elif y == 2000:
            val = 44.94
        elif y == 2001:
            val = 47.19
        elif y == 2002:
            val = 48.16
        elif y == 2003:
            val = 46.58
        elif y == 2004:
            val = 45.32
        elif y == 2005:
            val = 44.1
        elif y == 2006:
            val = 45.31
        elif y == 2007:
            val = 41.35
        elif y == 2008:
            val = 43.35
        elif y == 2009:
            val = 48.41
        elif y == 2010:
            val = 45.73
        elif y == 2011:
            val = 46.67
        elif y == 2012:
            val = 53.44
        elif y == 2013:
            val = 56.57
        elif y == 2014:
            val = 62.33
        elif y == 2015:
            val = 62.97
        elif y == 2016:
            val = 66.46
        elif y == 2017:
            val = 67.79
        elif y == 2018:
            val = 70.09
        elif y == 2019:
            val = 70.39
        elif y == 2020:
            val = 76.38
        elif y == 2021:
            val = 74.57
        elif y == 2022:
            val = 81.35
        elif y == 2023:
            val = 81.94
        else:
            val = 84.83

        row['Value'] = row['Value'] * val
        row["Element"]="Producer Price (SLC/tonne)"
        df["Flag Description"]="Official figure"
    return row


df = df.apply(convert_usd_to_slc, axis=1)


df.to_csv('converted_file.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X=df[["Element","Item","Year Code","Year","Unit","Flag"]]
Y=df["Value"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

**Linear Regression**

```python

# Preprocessing categorical columns with OneHotEncoding
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['Element', 'Item', 'Unit', 'Flag']),  # Categorical features
        ('num', 'passthrough', ['Year Code', 'Year'])  # Numerical features (pass through)
    ])

# Create a pipeline with preprocessing and linear regression
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Split data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Train the model
pipeline.fit(X_train, y_train)

# Make predictions on the test set
predictions = pipeline.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)  # Mean Squared Error
r2 = r2_score(y_test, predictions)  # R-squared score

# print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R): {r2*(100)}")
```

**Random Forest Implementation**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
# Preprocessing categorical columns with OneHotEncoding
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['Element', 'Item', 'Unit', 'Flag']),  # Categorical features
        ('num', 'passthrough', ['Year Code', 'Year'])  # Numerical features (pass through)
    ])

# Create a pipeline with preprocessing and Random Forest regression
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))  # RandomForestRegressor
])

# Split data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=42)

# Train the model
pipeline.fit(X_train, y_train)

# Make predictions on the test set
predictions = pipeline.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)  # Mean Squared Error
r2 = r2_score(y_test, predictions)  # R-squared score

# print(f"Predictions: {predictions}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R): {r2*(100)}")
```

```python
import pandas as pd
import zipfile

# Path to the zip file
zip_file_path = 'crop_production.csv.zip'

# Open the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # List all files in the zip archive
    file_list = zip_ref.namelist()
    print("Files in the zip archive:", file_list)

    # Assuming the CSV file is named 'crop_production.csv'
    csv_file_name = 'crop_production.csv'

    # Check if the CSV file exists in the zip archive
    if csv_file_name in file_list:
        # Extract the CSV file (optional, if you want to save it locally)
        zip_ref.extract(csv_file_name)

        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(csv_file_name)

        # Display the first few rows of the DataFrame
        print("Data from the CSV file:")
    else:
        print(f"File '{csv_file_name}' not found in the zip archive.")
df.head()
```

```python
import pandas as pd

# Assuming your data is already loaded into a DataFrame called `df`
# Example structure of `df`:
#    Year   Crop  Production
# 0  2020  Wheat         100
# 1  2020   Rice         150
# 2  2021  Wheat         120
# 3  2021   Rice         160
# 4  2022  Wheat         130
# 5  2022   Rice         170

# Group by 'Year' and 'Crop' and calculate the sum of production for each group
new_df = df.groupby(['Crop_Year', 'Crop'], as_index=False)['Production'].sum()

# Rename the columns for clarity
new_df.columns = ['Crop_Year', 'Crop', 'Sum of Production']

# Display the new dataset
print("New Dataset:")
print(new_df)

# Save the new dataset to a CSV file (optional)
new_df.to_csv('sum_of_production_by_year_and_crop.csv', index=False)
print("Dataset saved to 'sum_of_production_by_year_and_crop.csv'")
```
